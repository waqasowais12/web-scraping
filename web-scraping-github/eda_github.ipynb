## **EDA ON GITHUB REPOS (MACHINE LEARNING) USING PANDAS AND MATPLOTLIB**

### ***Import required libraries***

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### ***Load the dataset***

```python
df = pd.read_csv('./github_data.csv')
df.head()
```

### ***Data Understanding***

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

### ***Data Cleaning***

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
```

```python
df[df['Forks'].isnull()]
```

```python
df[df['Description'].isnull()]
```

```python
df = df.dropna(subset=['Description', 'Forks'])
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['Last Updated'])
```

```python
df.isnull().sum()
```

```python
df.shape
```

### ***EDA (Exploratory Analysis)***

- ***Top 10 languages Used***

```python
plt.figure(figsize=(10,5))
sns.countplot(y='Language', data=df, order=df['Language'].value_counts().index[:10])
plt.title('Top 10 languages Used in ML Repositories')
plt.show()
```

***Repositories with Most Stars***

```python
top_starred = df.sort_values(by='Stars', ascending=False).head(10)
plt.figure(figsize=(10,5))
sns.barplot(x='Stars', y='Repository', data=top_starred)
plt.title("Top 10 Most Starred Machine Learning Repositories")
plt.show()
```

- ***Correlation between Stars and Forks***

```python
sns.scatterplot(x='Stars', y='Forks', data=df)
plt.title('Stars vs Forks')
plt.show()
```

- ***Update Frequency*** 

```python
df['Year'] = df['Last Updated'].dt.year
plt.figure(figsize=(10,6))
sns.countplot(x='Year', data=df)
plt.title("Repository Update Frequency by Year")
plt.show()
```
